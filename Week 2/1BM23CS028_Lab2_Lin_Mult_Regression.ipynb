```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv("/content/canada_per_capita_income.csv")
df.info()
```

```python
display(df.isnull().sum())
```

```python
num_rows = len(df)
ones_column = np.ones((num_rows, 1))
year_column = df[['year']].values
matrix = np.column_stack((ones_column, year_column))

print("Shape of the new matrix:", matrix.shape)
```

```python
from numpy.linalg import inv
matrix_transpose = matrix.T
result_matrix = np.dot(matrix_transpose, matrix)
inverse_result_matrix = inv(result_matrix)
middle_result = np.dot(inverse_result_matrix, matrix_transpose)
final_result = np.dot(middle_result, df['per capita income (US$)'])
linear_reg_results = list(final_result)
print("Linear Regression Results: ", linear_reg_results)
year = int(input("Enter a year: \n"))
predicted_income = linear_reg_results[0] + linear_reg_results[1] * year
print("Predicted income for the year", year, "= ", predicted_income, "$")
```

```python
df1 = pd.read_csv("/content/salary.csv")
df1.info()
```

```python
df1['YearsExperience'] = df1['YearsExperience'].fillna(df1['YearsExperience'].mean())
df1.info()
```

```python
num_rows1 = len(df1)
ones_column = np.ones((num_rows1, 1))
year_column = df1[['YearsExperience']].values
matrix1 = np.column_stack((ones_column, year_column))
from numpy.linalg import inv
matrix_transpose1 = matrix1.T
result_matrix1 = np.dot(matrix_transpose1, matrix1)
inverse_result_matrix1 = inv(result_matrix1)
middle_result1 = np.dot(inverse_result_matrix1, matrix_transpose1)
final_result1 = np.dot(middle_result1, df1['Salary'])
linear_reg_results = list(final_result1)
print("Linear Regression Results: ", linear_reg_results)
year = int(input("Enter years of experience: \n"))
predicted_income = linear_reg_results[0] + linear_reg_results[1] * year
print("Predicted salary for years of experience", year, "= ", predicted_income)
```

```python
df2 = pd.read_csv("/content/hiring.csv")
```

```python
df2.info()
```

```python
experience_mapping = {'zero':0, 'one':1, 'two':2, 'three':3, 'four':4, 'five':5, 'six':6, 'seven':7, 'eight':8, 'nine':10, 'ten':10, 'eleven':11, 'twelve':12}
df2['experience'] = df2['experience'].map(experience_mapping)
imputer = SimpleImputer(strategy='mean')
df2['test_score(out of 10)'] = imputer.fit_transform(df2[['test_score(out of 10)']])
df2['experience'] = imputer.fit_transform(df2[['experience']])
df2.info()
```

```python
num_rows = len(df2)
ones_column = np.ones((num_rows, 1))
exp_col = df2[['experience']].values
test_col = df2[['test_score(out of 10)']].values
int_col = df2[['interview_score(out of 10)']].values
matrix2 = np.column_stack((ones_column, exp_col, test_col, int_col))
from numpy.linalg import inv
matrix_transpose1 = matrix2.T
result_matrix1 = np.dot(matrix_transpose1, matrix2)
inverse_result_matrix1 = inv(result_matrix1)
middle_result1 = np.dot(inverse_result_matrix1, matrix_transpose1)
final_result1 = np.dot(middle_result1, df2['salary($)'])
linear_reg_results = list(final_result1)
print("Linear Regression Results: ", linear_reg_results)
year = int(input("Enter years of experience: \n"))
test = int(input("Enter test score: \n"))
interview = int(input("Enter interview score: \n"))
predicted_income = linear_reg_results[0] + linear_reg_results[1] * year + linear_reg_results[2] * test + linear_reg_results[3] * interview
print("Predicted Salary for years of experience ", year, "test score ", test, "and interview_score ", interview, "= ", predicted_income)

```

```python
df3 = pd.read_csv("/content/1000_companies.csv")
```

```python
df3.info()
```

```python
from sklearn.preprocessing import OrdinalEncoder
encoder = OrdinalEncoder()
non_numeric_col_names = df3.select_dtypes(include='object').columns
df3[non_numeric_col_names] = encoder.fit_transform(df3[non_numeric_col_names])
df3.info()
```

```python
num_rows = len(df3)
ones_column = np.ones((num_rows, 1))
r_d_spend = df3[['R&D Spend']].values
admin = df3[['Administration']].values
market = df3[['Marketing Spend']].values
matrix2 = np.column_stack((ones_column, r_d_spend, admin, market))
from numpy.linalg import inv
matrix_transpose1 = matrix2.T
result_matrix1 = np.dot(matrix_transpose1, matrix2)
inverse_result_matrix1 = inv(result_matrix1)
middle_result1 = np.dot(inverse_result_matrix1, matrix_transpose1)
final_result1 = np.dot(middle_result1, df3['Profit'])
linear_reg_results = list(final_result1)
print("Linear Regression Results: ", linear_reg_results)
year = np.float64(input("Enter R and D spend: \n"))
test = np.float64(input("Enter Administration: \n"))
interview = np.float64(input("Enter Marketing Spend: \n"))
predicted_income = linear_reg_results[0] + linear_reg_results[1] * year + linear_reg_results[2] * test + linear_reg_results[3] * interview
print("Predicted Salary for R and D spend ", year, "Administration ", test, "and Marketing Spend ", interview, "= ", predicted_income)

```
